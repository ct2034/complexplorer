```python
import complexplorer as cx
import matplotlib.pyplot as plt
from math import sqrt
plt.rcParams["figure.figsize"] = (9, 9)
```

# Closed form Fibonacci

src https://www.youtube.com/watch?v=ghxQA3vvhsk&t=1069s

$$
\frac{(\phi)^n-(\frac{-1}{\phi})^n}{\sqrt{5}}
$$

with the golden ratio $\phi = \frac{1+\sqrt5}{2}$.



```python
domain = cx.Rectangle(12, 12)
phi = (1 + sqrt(5))/2
fib_func = lambda z: ((phi**z) - ((-1/phi)**z)) / sqrt(5)
cx.plot(domain, fib_func, n=1000)
plt.plot([0, 1, 2, 3, 5], [0, 0, 0, 0, 0], 'xk')
```

```python
domain = cx.Rectangle(100, 100)
cx.plot(domain, fib_func, n=1000)
```
